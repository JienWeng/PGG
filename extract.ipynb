```python
import pandas as pd
import glob
import os

# Define the r values you're interested in
r_values = [round(x * 0.5, 1) for x in range(3, 9)]  # gives [1.5, 2.0, ..., 4.0]

# Initialize empty lists to collect DataFrames
q_learning_dfs = []
double_q_learning_dfs = []

for r in r_values:
    folder = f"results_analysis/r{r}"
    q_path = os.path.join(folder, f"q_r{r}_metrics.csv")
    double_q_path = os.path.join(folder, f"double_q_r{r}_metrics.csv")

    # Read Q-learning metrics
    if os.path.exists(q_path):
        q_df = pd.read_csv(q_path)
        q_df['method'] = 'Q-learning'
        q_df['r'] = r
        q_learning_dfs.append(q_df)

    # Read Double Q-learning metrics
    if os.path.exists(double_q_path):
        dq_df = pd.read_csv(double_q_path)
        dq_df['method'] = 'Double Q-learning'
        dq_df['r'] = r
        double_q_learning_dfs.append(dq_df)

# Combine all data into single DataFrames
q_learning_all = pd.concat(q_learning_dfs, ignore_index=True)
double_q_learning_all = pd.concat(double_q_learning_dfs, ignore_index=True)

# Optionally combine both into a single DataFrame
all_data = pd.concat([q_learning_all, double_q_learning_all], ignore_index=True)
```

```python
all_data
```

```python
# Generate episode indices starting from 99 to 9999, step of 100
target_episodes = list(range(99, 10000, 100))

# Filter rows with episode in that list
filtered_data = all_data[all_data['episode'].isin(target_episodes)]

# Select the columns you care about
selected_columns = ['episode', 'avg_contribution', 
                    'agent_0_contrib', 'agent_1_contrib', 
                    'agent_2_contrib', 'agent_3_contrib', 
                    'method', 'r']

filtered_data = filtered_data[selected_columns]

filtered_data
```
